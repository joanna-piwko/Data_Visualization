```python
room_type = df_london_weekdays.room_type.value_counts()
```

## Bar plots

```python
plt.pie(room_type.values, 
        labels=room_type.keys())
        # autopct='%1.1f%%') - # print percentage of each label with specific accuracy
    
```

### Put pattern of each slice

```python
# list of all hatches -> https://matplotlib.org/stable/gallery/shapes_and_collections/hatch_style_reference.html

plt.pie(room_type.values, 
        labels=room_type.keys(),
        hatch=['x', 'O',  '*'],
        autopct='%1.1f%%')  # print percentage of each label with specific accuracy
    
```

```python

plt.pie(room_type.values, 
        labels=room_type.keys(),
        autopct='%1.1f%%',
        pctdistance=1.3, 
        labeldistance=.3)  # print percentage of each label with specific accuracy
    
```

### Take out - "explode" slice from pie chart

```python
explode = (0, 0, 0.5)
plt.pie(room_type.values, 
        labels=room_type.keys(),
        autopct='%1.1f%%',
        explode=explode)
```

```python
explode = (0, 0, 0.5)
plt.pie(room_type.values, 
        labels=room_type.keys(),
        autopct='%1.1f%%',
        explode=explode,
        shadow=True, 
        startangle=90)
```

### Adding frame, setting radius of plot

```python
colors = plt.get_cmap('Blues')(np.linspace(0.5, 0.9, len(room_type.values)))

plt.pie(room_type.values, 
        labels=room_type.keys(),
        autopct='%1.1f%%',
        colors=colors, 
        radius=3, center=(4, 4),
#        wedgeprops={"linewidth": 1, "edgecolor": "white"},
        frame=True)
```

```python
fig, ax = plt.subplots()
ax.pie(room_type.values, 
        labels=room_type.keys(),
        autopct='%1.1f%%',
        radius=3, center=(4, 4),
        wedgeprops={"linewidth": 1, "edgecolor": "white"},
        frame=True)

ax.set(xlim=(0, 15), xticks=np.arange(1, 15),
       ylim=(0, 8), yticks=np.arange(1, 8))
# plt.tick_params(top=False, bottom=False, left=False, right=False,
#                 labelleft=False, labelbottom=False)
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
```

```python
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(room_type.values, 
        labels=room_type.keys(),
        autopct='%1.1f%%')

ax.legend(wedges, room_type.keys(),
          title="Rome types",
          loc="best")
```

```python
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(room_type.values, 
        labels=room_type.keys(),
        autopct='%1.1f%%')

ax.legend(wedges, room_type.keys(),
          title="Rome types",
          loc="center left",
          bbox_to_anchor=(1.5, 0, 0.5, 1)) #top right-hand corner - x, y, width, height
```

```python
x = np.arange(14)
y = np.sin(x / 2)

plt.step(x, y, where='post', label='post')
plt.plot(x, y, 'o--', color='grey', alpha=0.3)

plt.grid(axis='x', color='0.95')
plt.legend(title='Parameter where:')
plt.title('plt.step(where=...)')
plt.show()
```

```python
df = pd.read_csv("C:\\Users\\asiun\\PycharmProjects\\exoskeletons_research\\laboratories\\Data_Visualization\\data\\UNdata_Export_20230404_122538843.csv")
```

```python
df["Country or Area"].unique()
```

```python
fig, ax = plt.subplots(1, figsize=(10,10))

countries = df["Country or Area"].unique()

for i in countries:
    # get a single country from the list
    temp = df[df['Country or Area'] == i]
    # plot the lines
    plt.plot(temp.Year, temp.Value, marker='o', markersize=5)
    # replace large labels
    if i == 'China, Macao Special Administrative Region':
        i = 'Macao (China)'
    # end label
    plt.text(temp.Year.values[0]+0.02, temp.Value.values[0], i)
    # start label
    plt.text(temp.Year.values[1]-0.02, temp.Value.values[1], i, ha='right')
    
# x limits, x ticks, and y label 
plt.xlim(2017.5,2019.5)
plt.xticks([2018, 2019])
# get y ticks, replace 1,000 with k, and draw the ticks
yticks = plt.yticks()[0]
plt.yticks(yticks, ['{}k'.format(i/1000) for i in yticks])
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))



# generate some random test data
all_data = [df_london_weekdays.dist.values, df_london_weekdays.metro_dist.values]

# plot violin plot
axs[0].violinplot(all_data,
                  showmeans=False,
                  showmedians=True)
axs[0].set_title('Violin plot')

# plot box plot
axs[1].boxplot(all_data)
axs[1].set_title('Box plot')

# adding horizontal grid lines
for ax in axs:
    ax.yaxis.grid(True)
    ax.set_xticks([1, 2],
                  labels=['Distance to city center', 'Distance to metro station'])
    ax.set_ylabel('Observed values')

plt.show()
```
